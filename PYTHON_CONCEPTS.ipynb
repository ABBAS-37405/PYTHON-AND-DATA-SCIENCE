**Pool of String Literals in Python**
=====================================

In Python, a pool of string literals is a mechanism used by the interpreter to manage memory allocation for string objects. When a string literal is encountered in the code, Python checks if an identical string already exists in memory. If it does, the new string is not created, and instead, a reference to the existing string is used.

**How it Works**
---------------

When Python encounters a string literal, it checks the following:

1. **`Interning`**: Python checks if the string is already interned, meaning it has been created and stored in memory before. If it is, Python returns a reference to the existing string.
2. **`String Literal Pool`**: If the string is not interned, Python checks the string literal pool, which is a cache(Pronunciation: `kash`) of recently created strings. If the string is found in the pool, Python returns a reference to the existing string.
3. **`Create New String`**: If the string is not found in the pool or interned, Python creates a new string object and stores it in memory.

**Benefits**
------------

The pool of string literals provides several benefits:

* **`Memory Efficiency`**: By reusing existing strings, Python reduces memory allocation and deallocation overhead.
* **`Performance`**: Looking up strings in the pool is faster than creating new strings.
* **`Internalization`**: The pool helps to internalize strings, making them more efficient to use in comparisons and other operations.


## **Interning in Python**

### **1.  What is Interning?**

Interning is a process where Python stores only one copy of certain strings in memory and reuses them whenever the same string is used again. This is done to save memory and speed up comparisons.

### **2. Which Strings are Interned?**

* **Short strings**: Usually strings that are **`20 characters or fewer`**.
* **Identifiers**: Strings that look like variable names (e.g., "x", "my_var").
* **Explicit interning**: You can manually intern a string using the sys.intern() function.

````python
class Employee:
  def __init__(self, name, salary, project):
    self.name=name
    self.salary=salary
    self.project=project
  def show(self):
    print(f"Employee Name: {self.name}")
    print(f"Employee Salary: {self.salary}")
    print(f"Employee Project: {self.project}")
emp_name=input("Enter Employee Name: ")
emp_salary=int(input("Enter Employee Salary: "))
emp_project=input("Enter Employee Project: ")
emp=Employee(emp_name,emp_salary,emp_project)
emp.show()
````

````python
class BankAccount:
  def __init__(self,name,account_number,balance=0,min=300):
    self.name=name
    self.account_number=account_number
    self.balance=balance
    self.min=min
  def deposit(self,amount):
    self.balance+=amount
    print(f"Your new balance is {self.balance +self.min}")
  def withdraw(self,amount):
    if self.balance-amount>=self.min:
      self.balance-=amount
      print(f"Your new balance is {self.balance+self.min}")
    else:
      print("Insufficient balance")
  def showbalance(self):
    print(f"Your balance is {self.balance + self.min}")
flag=True
while flag:
  name=input("Enter your name: ")
  account_number=input("Enter your account number: ")
  acc=BankAccount(name,account_number)
  acc.deposit(int(input("Enter the amount to deposit: ")))
  acc.withdraw(int(input("Enter the amount to withdraw: ")))
  acc.showbalance()
  print("Do you want to continue? (y/n)")
  if input()!="y":
    flag=False
````

# LINEAR REGRESSION

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# Sample data with multiple features
# Let's assume we are predicting the price of a house based on size and number of bedrooms
data = {'size': [50, 60, 70, 80, 90, 100, 55, 65, 75, 85, 95, 105],
        'bedrooms': [1, 1, 2, 2, 3, 3, 1, 2, 2, 3, 3, 4],
        'price': [200000, 250000, 300000, 350000, 400000, 450000, 220000, 280000, 330000, 380000, 430000, 480000]}
df = pd.DataFrame(data)

# Features (X) and target (y)
X = df[['size', 'bedrooms']]
y = df['price']

# Create a multiple linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict the price for a new house with size 75 sqft and 2 bedrooms
new_house = np.array([[75, 2]])
predicted_price = model.predict(new_house)

print(f"Predicted price for a 75 sqft house with 2 bedrooms: ${predicted_price[0]:,.2f}")

# Visualize the results (3D scatter plot for two features)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['size'], df['bedrooms'], df['price'], c='blue', marker='o')

# Create a meshgrid for the plane
size_surf, bedrooms_surf = np.meshgrid(np.linspace(df['size'].min(), df['size'].max(), 100),
                                       np.linspace(df['bedrooms'].min(), df['bedrooms'].max(), 100))
price_surf = model.predict(np.c_[size_surf.ravel(), bedrooms_surf.ravel()]).reshape(size_surf.shape)

ax.plot_surface(size_surf, bedrooms_surf, price_surf, color='red', alpha=0.5)

ax.set_xlabel('House Size (sqft)')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('House Price (USD)')
ax.set_title('Multiple Linear Regression Example')

plt.show()
````
